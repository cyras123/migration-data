### Geographic Data Generator

```python
import json as json
import numpy as np
import random as random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### Cities

```python
citydf = pd.read_csv('cities_data_4.csv',dtype = str)
citydf.head()
```

```python
raw_country = citydf["Country or Area"]
raw_city = citydf["City"]
```

```python
unique_countries = list(set(raw_country))

cities = dict()

for ucn in unique_countries:
    
    allcities = []
    
    for cn,ct in zip(raw_country,raw_city):
        if cn == ucn:
            allcities.append(str(ct))
            
    cities.update({str(ucn):list(set(allcities))})
```

```python
with open("cities.json", 'w') as f:
    json.dump(cities,f)
```

#### Lat-Long

```python
with open("countries_lat_long.json", 'r') as fw:
    latlong = json.load(fw)
```

```python
latlong[0:5]
```

#### Countries-cities-lat-long

```python
Data = []
for item in latlong:
    country = item['name']
    ID = item['sym']
    lat = item['lat']
    lon = item['lon']
    ctz = "NA"
    for cn in cities.keys():
        if cn == country:
            ctz = cities[cn]
        
            
    Data.append({'name': country, \
                 "cities":ctz,\
                 "lat": lat,\
                 "lon":lon,\
                  'ID': ID})    
    
```

```python
with open("world.json", 'w')as fw:
    json.dump(Data,fw)
```
